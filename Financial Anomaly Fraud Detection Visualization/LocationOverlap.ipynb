```python
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shp_path = "/content/MC2/MC2/Geospatial/Kronos_Island.shp'"
sf = shp.Reader(shp_path)
```

```python
def read_shapefile(sf):
    """
    Read a shapefile into a Pandas dataframe with a 'coords'
    column holding the geometry information. This uses the pyshp
    package
    """
    fields = [x[0] for x in sf.fields][1:]
    records = sf.records()
    shps = [s.points for s in sf.shapes()]
    df = pd.DataFrame(columns=fields, data=records)
    df = df.assign(coords=shps)
    return df
```

```python
df = read_shapefile(sf)
df.shape
```

```python
import geopandas as gpd
import rasterio
from rasterio.mask import mask
import matplotlib.pyplot as plt

# Define paths to your image and shapefile
img_path = '/content/MC2/MC2/MC2-tourist.jpg'  # Replace with the path to your JPEG image file
shapefile_path = '/content/MC2/MC2/Geospatial/Kronos_Island.shp'  # Replace with the path to your shapefile

# Read the image using rasterio
img_dataset = rasterio.open(img_path)

# Load your shapefile
shapefile = gpd.read_file(shapefile_path)

# Reproject the shapefile to match the CRS of the image
shapefile = shapefile.to_crs(crs=img_dataset.crs)

# Mask the image using the shapefile's geometry
masked_image, _ = mask(img_dataset, shapes=shapefile.geometry, crop=True)

# Plot the masked image
plt.figure(figsize=(10, 10))
plt.imshow(masked_image.transpose(1, 2, 0))  # Transpose the image array to match the correct format
shapefile.plot(ax=plt.gca(), color='red', alpha=0.5)  # Adjust color and transparency as needed
plt.show()
```
